# Training a radom forest model

```python
#Import stuff
import numpy as np
import pandas as pd
from sklearn.ensemble import RandomForestRegressor
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error
```

```python
#Import data
data = pd.read_csv("merged_data.csv")
data.head()
```

```python
#Drop some columns that aren't needed
#This data is theoretically already represented by store_nbr
data = data.drop(columns=["city", "state", "type", "cluster"])
data.head()
```

```python
#Converting categorical values using integer encoding
products = data.family.values
products = list(set(products))
products_dict = dict()
for i in range(len(products)):
    products_dict[products[i]] = i+1
    
data.family = data.family.map(products_dict)
```

```python
#Splitting the data
train = data[data["date"] > "2014-12-31"]
train = train[train["date"] < "2017-01-01"]

test = data[data["date"] > "2016-12-31"]

#Also dropping id and date as these aren't needed
train = train.drop(columns=["id", "date"])
test = test.drop(columns=["id", "date"])

X_train = train.drop(columns="sales")
y_train = train["sales"]

X_test = test.drop(columns="sales")
y_test = test["sales"]
```

```python
#Train the model
rf = RandomForestRegressor(random_state=1).fit(X_train, y_train)
```

```python
#Making predictions
y_pred = rf.predict(X_test)
```

```python
# Evaluate the model
mse = mean_squared_error(y_test, y_pred)
rmse = np.sqrt(mse)
print(f'Root Mean Squared Error: {rmse}')
```
